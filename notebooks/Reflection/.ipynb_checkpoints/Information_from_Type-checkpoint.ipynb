```
using System;
using System.Reflection;
using System.Linq;
```

```
Type t = typeof(string);

Console.WriteLine("Assembly: {0}", t.Assembly);
Console.WriteLine("Name: {0}", t.Name);
Console.WriteLine("NameSpace: {0}", t.Namespace);
Console.WriteLine("Attributes: {0}", t.Attributes);
Console.WriteLine("\nstring Method names:");

MethodInfo[] methods = t.GetMethods();

foreach (var item in methods)
{
  Console.WriteLine("\nMethod Name: {0}", item.Name);

  ParameterInfo[] inputparam = item.GetParameters();
  inputparam.ToList().ForEach(x => Console.WriteLine("\tParameter: {0}  Type: {1}", x.Name, x.ParameterType));
}
```
