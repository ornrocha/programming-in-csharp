# Consume xml

```
using System;
using System.Collections.Generic;
using System.IO;
using System.Xml;
using System.Xml.Serialization;
```

```
string xmlpath = "Data/employees.xml";
```

```
public static string GetEmployeesString()
{
  string xmlpath = "Data/employees.xml";
  FileStream fs = new FileStream(xmlpath, FileMode.Open, FileAccess.Read);

  string contents;
  using (var sr = new StreamReader(fs))
  {
     contents = sr.ReadToEnd();
  }
  return contents;
}
```

## Read using StringReader $\to$ that accepts a string with content

```
using (StringReader strreader = new StringReader(GetEmployeesString()))
{
    XmlReader xReader = XmlReader.Create(strreader);

    while (xReader.Read())
    {
       Console.WriteLine("{0}\t{1}", xReader.Name, xReader.Value);
    }
}
```

### Using XmlTextReader

```
using (StringReader strreader = new StringReader(GetEmployeesString()))
{
    XmlTextReader reader = new XmlTextReader(strreader);

    while (reader.Read())
    {
       Console.WriteLine("{0}\t{1}", reader.Name, reader.Value);
    }
}
```

## XmlDocument

```
string xmlpath = "Data/employees.xml";

XmlDocument xmldoc = new XmlDocument();

FileStream fs = new FileStream(xmlpath, FileMode.Open, FileAccess.Read);
xmldoc.Load(fs);
```

```
XmlNodeList nodeList = xmldoc.DocumentElement.SelectNodes("/ListEmployees/Employees/Employee");

foreach (XmlNode node in nodeList)
{
   Console.WriteLine(node.SelectSingleNode("emailAddress").InnerText);
}
```

```
XmlNodeList xmlnode = xmldoc.GetElementsByTagName("Employee");

for (int i = 0; i < xmlnode.Count; i++)
{
    Console.WriteLine("###### Node {0} ######", i+1);
    for (int j = 0; j < xmlnode[i].ChildNodes.Count; j++)
    {
       Console.WriteLine(xmlnode[i].ChildNodes.Item(j).InnerText);
    }

     Console.WriteLine("################\n\n");
}
```
